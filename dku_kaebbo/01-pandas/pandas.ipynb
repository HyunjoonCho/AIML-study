### 시각화
다양하게 그려보며 추세를 파악  
데이터가 너무 많으면 1) 보기 어렵고 2) 속도가 느림 -> pandas에서 제공하는 sample 함수를 활용  
* 아래 그래프들이 어떻게 그려지는가?  
  pairplot -   자세한 내용은 [document](https://seaborn.pydata.org/generated/seaborn.pairplot.html) 참고  
  vars에 주어진 변수들의 모든 쌍에 대해 상관관계를 그려줌, hue는 추가로 구분해주는 용도(여기서는 종속 변수)  
* 각 그래프에서 어떤 정보를 알 수 있는가?  
  분포가 다르면 다를수록 클래스 변수를 예측하는데 유용하다

### 시험 데이터 로드

```python
tst = pd.read_csv(tst_file, index_col=0)
print(tst.shape)
tst.head() # row 수는 25%, column은 class 하나 빠져서 18개
```

### 학습/시험 데이터 결합 및 피쳐 조작

```python
df = pd.concat([trn, tst], axis=0) # df = data frame
# axis는 결합 방향 0 - 세로 결합, 1 - 가로 결합
print(df.shape)
df.tail() # column 수가 많은 train을 기준으로 shape이 결정됨, 이때 test set의 class column은 NaN
```

```python
df.fillna(-1, inplace=True)
df.tail()
```

```python
df['nObserve'].hist(bins=30)
```

```python
df['nObserve'] = df['nObserve'].apply(np.log1p) # inverse 연산은 expm1
# log1p = (1을 더한 후에 로그를 취함); -inf가 나올 수 있으므로
# 멱함수 형태의 데이터는 선형 회귀가 적용이 어려움, 이때 흔히 가하는 조작이 log1p
```

```python
df['nObserve'].hist(bins=30)
```

```python
df['d_dered_u'] = df['dered_u'] - df['u'] # feature 추가는 일반적인 python 문법과 크게 다르지 않음
df['d_dered_g'] = df['dered_g'] - df['g']
df['d_dered_r'] = df['dered_r'] - df['r']
df['d_dered_i'] = df['dered_i'] - df['i']
df['d_dered_z'] = df['dered_z'] - df['z']
df['d_dered_rg'] = df['dered_r'] - df['dered_g']
df['d_dered_ig'] = df['dered_i'] - df['dered_g']
df['d_dered_zg'] = df['dered_z'] - df['dered_g']
df['d_dered_ri'] = df['dered_r'] - df['dered_i']
df['d_dered_rz'] = df['dered_r'] - df['dered_z']
df['d_dered_iz'] = df['dered_i'] - df['dered_z']
df['d_obs_det'] = df['nObserve'] - df['nDetect']
print(df.shape)
df.head()
```

```python
df.corr().style.background_gradient() # correlation을 보여줌
```

```python
df.drop(['airmass_z', 'airmass_i', 'airmass_r', 'airmass_g', 'u', 'g', 'r', 'i', 'nDetect', 'd_dered_rg', 'd_dered_ri'], 
        axis=1, inplace=True) 
# 왜 하필 이들을 삭제할까? 중복되는 column은 삭제하는 것이 좋다!
# airmass 끼리 corr이 높고 앞선 barh에서도 비슷한 결과 -> 네 개 삭제 후 하나만 남김
# u - dered_u, g - dered_g 등의 corr이 높음 -> 한 쪽 삭제
print(df.shape)
df.head()
```

```python
df.to_csv(feature_file)
```
